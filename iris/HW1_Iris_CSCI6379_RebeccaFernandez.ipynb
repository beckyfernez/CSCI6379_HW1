```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.preprocessing import OneHotEncoder

# Load dataset
iris = fetch_ucirepo("iris")
X = iris.data.features
y = iris.data.targets

# Flatten y (numpy array)
y = y.to_numpy().flatten()

# Shuffle dataset
X, y = shuffle(X, y, random_state=42)

# Convert classes to binary code (One-Hot Encoding)
encoder = OneHotEncoder(sparse_output=False)
Y = encoder.fit_transform(y.reshape(-1, 1))
#print("First few transformed Y values:\n", Y[:5])

# Split data (training: 80%, test: 20%)
n_samples = X.shape[0]
n_train = int(0.8 * n_samples)

trainX, trainY = X[:n_train], Y[:n_train]
testX, testY = X[n_train:], Y[n_train:]

# Reference slide deck "2-11_Multiclass_Classification", slide 10
trainX = np.array(trainX)
trainY = np.array(trainY)
testX = np.array(testX)
testY = np.array(testY)

# Initialize Parameters
num_features = trainX.shape[1]  # 4 features (Sepal Length: 0, Sepal Width: 1, Petal Length: 2, and Petal Width:3)
num_classes = trainY.shape[1]   # 3 (one-hot encoded)

# Reference slide deck "2-11_Multiclass_Classification", slide 10
W = np.zeros((num_features, num_classes))
B = np.zeros(num_classes)
alpha = 0.005
N = trainX.shape[0]
C = W.shape[1]

# Gradient Descent
# Reference slide deck "2-11_Multiclass_Classification", slide 11
num_iterations = 2000
for i in range(num_iterations):
    for j in range(C):
        error = np.dot(trainX, W[:, j]) + B[j] - trainY[:, j]
        W[:, j] -= alpha * (1/N) * np.dot(trainX.T, error)
        B[j] -= alpha * (1/N) * np.sum(error)

    # Print W and B periodically
    if i % 200 == 0:
        print(f"Iteration {i}: W =\n{W}\nB = {B}")

# Model Evaluation: Predict and Calculate Error
def predict(X, W, B):
    scores = np.dot(X, W) + B
    return np.argmax(scores, axis=1)

train_pred = predict(trainX, W, B)
test_pred = predict(testX, W, B)

train_accuracy = np.mean(np.argmax(trainY, axis=1) == train_pred)
test_accuracy = np.mean(np.argmax(testY, axis=1) == test_pred)

train_error = 1 - train_accuracy
test_error = 1 - test_accuracy

print(f"Final Train Error: {train_error:.4f}, Final Test Error: {test_error:.4f}")


# Plot Training and Testing Errors
plt.figure(figsize=(8, 5))
plt.plot(range(num_iterations), [train_error]*num_iterations, 'b-', label="Training Error")
plt.plot(range(num_iterations), [test_error]*num_iterations, 'r-', label="Testing Error")
plt.xlabel("Iteration")
plt.ylabel("Error (1 - Accuracy)")
plt.title("Training and Testing Error vs. Iterations")
plt.legend()
plt.show()
```
